# Full Murex-Style Risk Pipeline
End-to-end simulation: data → pricing → Greeks → VaR → aggregation.

```

import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime, timedelta

VALUATION_DATE = datetime(2025,4,4)
tickers = ['AAPL','GOOG','EURUSD=X']

# Market Data
data = yf.download(tickers, start="2020-01-01", end="2025-04-04")['Close']
data.ffill(inplace=True)

# Returns
returns = np.log(data/data.shift(1)).dropna()
returns.head()
```

```

# Portfolio
weights = np.array([0.5,0.3,0.2])
portfolio_returns = returns.dot(weights)

# VaR
var_99 = np.percentile(portfolio_returns,1)
print("VaR 99%:", var_99)
```
